```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.barplot(x='time', y='total_bill', data=tips)
```

```python
sns.barplot(x='size', y='total_bill', data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='size', y='total_bill', data=tips[tips['time']=='Dinner'])
```

```python
sns.barplot(x='size', y='total_bill', data=tips[tips['time']=='Dinner'], estimator=np.std)
```

```python
sns.countplot(x='size', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='size')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.violinplot(x='tip', y='sex', data=tips, kind='violin')
```

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
flights.head()
```

```python
passengers_per_month = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(passengers_per_month, cmap='coolwarm', linewidth=1)
```

```python
tips.head()
```

```python
tips_by_day = tips.pivot_table(index='day', columns='size', values='total_bill')
```

```python
sns.heatmap(tips_by_day)
```

```python
sns.clustermap(passengers_per_month, cmap='coolwarm', standard_scale=1)
```
